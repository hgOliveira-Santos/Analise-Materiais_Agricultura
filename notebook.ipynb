```python
# Importando os pacotes necess√°rios
import numpy as np
import pandas as pd
```

# Carregando o conjunto de dados


```python
df = pd.read_csv('agricultural_raw_material.csv')
df
```

# Explorando os dados

```python
df.info
```

```python
# Verificando a exist√™ncia de valores nulos em cada coluna
df.isnull().sum()
```

# Tratando os dados

```python
# Substituindo '%', '-' e ','
df = df.replace('%', ',', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df
```

```python
# Removendo linhas com valores NaN
df = df.dropna()

# Verificando se todos os valores NaN foram tratados
df.isnull().sum()
```

```python
df
```

#### Convertendo o tipo de dado para float


```python
df.dtypes
```

Diversas colunas do dataframe possuem valores do tipo 'object'. Por√©m, √© necess√°rio transform√°-las em tipo de dado 'float', para a realiza√ß√£o de c√°lculos.

```python
# Convertendo de object para float
lst = list(df.columns)
lst.remove('Month')
lst
```

```python
df[lst] = df[lst].astype('float')
df.dtypes
```

```python
# Formatando a coluna datetime
df.Month = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)
```

```python
# Indexando a coluna Month
df = df.set_index('Month')
df.head(10)
```

# An√°lise Explorat√≥ria e Visualiza√ß√£o


```python
# Importando os pacotes necess√°rios
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Configura√ß√µes de visualiza√ß√£o dos gr√°ficos
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

## Mapas de calor

### Mapa de calor para os pre√ßos

```python
# Lista dos dados brutos -> Price
price = np.array([col for col in df.columns if '%' not in col])
price
```

```python
# Matriz de correla√ß√£o para price
corrmat_price = df[price].corr()
corrmat_price
```

```python
# Configurando tamanho do gr√°fico
plt.figure(figsize = (12, 9))

# Mascarando a parte superior da matriz, vide sua simetria
mask = np.triu(np.ones_like(corrmat_price, dtype=bool))

# Gr√°fico
sns.heatmap(corrmat_price, vmax=.8, mask=mask, square=True, annot=True)
plt.show()
```


### Mapa de calor para percentuais

```python
# Lista dos dados brutos -> Change
change = np.array([col for col in df.columns if '%' in col])
change
```

```python
# Matriz de correla√ß√£o para change
corrmat_change = df[change].corr()
corrmat_change
```

```python
# Configurando o tamanho do gr√°fico
plt.figure(figsize = (24, 15))

# Gr√°fico
sns.heatmap(corrmat_change, annot=True)
plt.show()
```

Com base no gr√°fico, pode-se concluir que quase n√£o h√° rela√ß√£o entre a varia√ß√£o percentual dos pre√ßos das mat√©rias-primas.

## L√£ grossa


```python
axes = df[['Coarse wool Price', 'Coarse wool price % Change']].plot(figsize=(11, 9), subplots = True, linewidth = 1)
```

## Perguntas e respostas



###**Q1: Qual a varia√ß√£o normal do pre√ßo de cada mat√©ria-prima?**


```python
for i in range(len(change)):
    plt.figure(figsize = (12, 12))
    df[change[i]].hist(figsize = (11, 9), linewidth = 1)
    plt.xlabel('% Change')
    plt.ylabel('count')
    plt.legend(change[i:], loc = 'upper center', bbox_to_anchor = (1.2, 1))
```

###**Q2: Qual mat√©ria-prima tem o menor pre√ßo ao longo dos anos?**

```python
plt.figure(figsize=(10, 10))
for i in range(len(price)):
    plt.subplot(4, 3, i+1)
    plt.subplots_adjust(hspace = 1, wspace = 0.5)
    plt.title(price[i])
    plt.plot(df[price[i]])
    plt.xticks(rotation=90)

plt.suptitle('Raw-Materials price comparision')
```

Atrav√©s dos gr√°ficos, percebe-se que o algod√£o (Cotton) e a borracha (Rubber) possu√≠ram os pre√ßos mais baixos ao longo do tempo.

#### Compara√ß√£o entre os pre√ßos do algod√£o e da borracha

```python
plt.figure(figsize=(10, 10))
plt.plot(df[['Cotton Price', 'Rubber Price']])
plt.title('Raw-Materials price comparision')
plt.xlabel('Years')
plt.ylabel('Prices')
plt.legend(['Cotton Price', 'Rubber Price'], loc = 'upper center', bbox_to_anchor=(1.2, 1))
```

```python
plt.figure(figsize=(10, 10))
plt.plot(df[price])
plt.title("Raw-Materials price comparision")
plt.xlabel('Years')
plt.ylabel('Prices')
plt.legend(price, loc='upper center',bbox_to_anchor=(1.2,1))
```

###**Q3: Qual mat√©ria-prima tem o maior e menor varia√ß√£o de porcentagem de pre√ßo?**


```python
import random as random
```

```python
plt.figure(figsize=(10, 10))
for i in range(len(change)):

    r = random.random()
    b = random.random()
    g = random.random()

    color = (r, g, b)

    plt.subplot(4, 3, i+1)
    plt.subplots_adjust(hspace = 1, wspace = 0.5)
    plt.plot(df[change[i]], c = color)
    plt.xticks(rotation = 90)
    plt.title(change[i])
    plt.xlabel('Years')
    plt.ylabel('& Change')
```

###**Q4: Encontrando as mat√©rias-primas com mudan√ßa dr√°stica de pre√ßo**


```python
lowlist = ['Cotton Price', 'Hide Price', 'Softlog Price', 'Hard log Price', 'Soft sawnwood Price', 'Rubber Price']
plt.figure(figsize=(12, 12))
plt.ylabel('Prices')
plt.xlabel('Years')
for i in range(len(lowlist)):
    sns.scatterplot(y=df[lowlist[i]], x = df.index)
    plt.legend(lowlist, loc = 'upper center', bbox_to_anchor = (1.2, 1))
```

```python
highlist = ['Coarse wool Price', 'Copra Price', 'Fine wool Price', 'Hard sawnwood Price', 'Plywood Price', 'Wood pulp Price']
plt.figure(figsize=(12, 12))
plt.ylabel('Prices')
plt.xlabel('Years')
for i in range(len(highlist)):
    sns.scatterplot(x=df[highlist[i]], y = df.index)
    plt.legend(highlist, loc = 'upper center', bbox_to_anchor = (1.2, 1))
```

Percebe-se uma mudan√ßa de pre√ßo dr√°stica para a tora dura entre as mat√©rias de baixa faixa de pre√ßo. Por outro lado, para as mat√©rias de alto pre√ßo, os pre√ßos da l√£ fina possuem tal varia√ß√£o.





###**Q5: Descobrindo a faixa de pre√ßo de mat√©rias-primas de baixo pre√ßo.**

```python
plt.figure(figsize=(10, 10))
for i in range(len(lowlist)):
    plt.subplot(2, 3, i+1)
    plt.subplots_adjust(hspace=0.5)
    sns.boxplot(x=df[lowlist[i]])
```
