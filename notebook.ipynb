```python
# Importando os pacotes necessários
import numpy as np
import pandas as pd
```

# Carregando o conjunto de dados


```python
df = pd.read_csv('agricultural_raw_material.csv')
df
```

# Explorando os dados

```python
df.info
```

```python
# Verificando a existência de valores nulos em cada coluna
df.isnull().sum()
```

# Tratando os dados

```python
# Substituindo '%', '-' e ','
df = df.replace('%', ',', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df
```

```python
# Removendo linhas com valores NaN
df = df.dropna()

# Verificando se todos os valores NaN foram tratados
df.isnull().sum()
```

```python
df
```

#### Convertendo o tipo de dado para float


```python
df.dtypes
```

Diversas colunas do dataframe possuem valores do tipo 'object'. Porém, é necessário transformá-las em tipo de dado 'float', para a realização de cálculos.

```python
# Convertendo de object para float
lst = list(df.columns)
lst.remove('Month')
lst
```

```python
df[lst] = df[lst].astype('float')
df.dtypes
```

```python
# Formatando a coluna datetime
df.Month = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)
```

```python
# Indexando a coluna Month
df = df.set_index('Month')
df.head(10)
```

# Análise Exploratória e Visualização


```python
# Importando os pacotes necessários
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Configurações de visualização dos gráficos
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

## Mapas de calor

### Mapa de calor para os preços

```python
# Lista dos dados brutos -> Price
price = np.array([col for col in df.columns if '%' not in col])
price
```

```python
# Matriz de correlação para price
corrmat_price = df[price].corr()
corrmat_price
```

```python
# Configurando tamanho do gráfico
plt.figure(figsize = (12, 9))

# Mascarando a parte superior da matriz, vide sua simetria
mask = np.triu(np.ones_like(corrmat_price, dtype=bool))

# Gráfico
sns.heatmap(corrmat_price, vmax=.8, mask=mask, square=True, annot=True)
plt.show()
```


### Mapa de calor para percentuais

```python
# Lista dos dados brutos -> Change
change = np.array([col for col in df.columns if '%' in col])
change
```

```python
# Matriz de correlação para change
corrmat_change = df[change].corr()
corrmat_change
```

```python
# Configurando o tamanho do gráfico
plt.figure(figsize = (24, 15))

# Gráfico
sns.heatmap(corrmat_change, annot=True)
plt.show()
```

Com base no gráfico, pode-se concluir que quase não há relação entre a variação percentual dos preços das matérias-primas.

## Perguntas e respostas



###**Q1: Qual a variação normal do preço de cada matéria-prima?**


```python
for i in range(len(change)):
    plt.figure(figsize = (12, 12))
    df[change[i]].hist(figsize = (11, 9), linewidth = 1)
    plt.xlabel('% Change')
    plt.ylabel('count')
    plt.legend(change[i:], loc = 'upper center', bbox_to_anchor = (1.2, 1))
```

###**Q2: Qual matéria-prima tem o menor preço ao longo dos anos?**

```python
plt.figure(figsize=(10, 10))
for i in range(len(price)):
    plt.subplot(4, 3, i+1)
    plt.subplots_adjust(hspace = 1, wspace = 0.5)
    plt.title(price[i])
    plt.plot(df[price[i]])
    plt.xticks(rotation=90)

plt.suptitle('Raw-Materials price comparision')
```

Através dos gráficos, percebe-se que o algodão (Cotton) e a borracha (Rubber) possuíram os preços mais baixos ao longo do tempo.

#### Comparação entre os preços do algodão e da borracha

```python
plt.figure(figsize=(10, 10))
plt.plot(df[['Cotton Price', 'Rubber Price']])
plt.title('Raw-Materials price comparision')
plt.xlabel('Years')
plt.ylabel('Prices')
plt.legend(['Cotton Price', 'Rubber Price'], loc = 'upper center', bbox_to_anchor=(1.2, 1))
```
